```python
# importation des bibliothèques

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# importation du fichier

data_presidentielles = pd.read_csv("elections-presidentielles-2017-1ertour.csv",  sep =';') # pour ouvrir le fichier csv
print(data_presidentielles)

data_presidentielles.columns    # pour voir les noms des colonnes
```

```python
# 1.voix pour JL MELENCHON

voix_JL = data_presidentielles[['MÉLENCHON Jean-Luc']]  # on extrait la colonne 'MÉLENCHON Jean-Luc'
display(voix_JL)    # 'display' permet d'afficher 
```

```python
# 2.représentatione en histogramme

voix_JL.hist(bins = 20) # histogramme avec 'nb de voix' en abscisses et, par 'intervalle de voix [x, x + dx]', le 'nb d'éléments qui se trouvent dans [x, x + dx]' en ordpnnées 
```

```python
# 3.nb. de bureaux de vote par arrondissement

data_ar = data_presidentielles[['NUM_ARROND']]  # les bureaux de vote étant une clé primaire de la base de données (ie. on peut identifier chaque ligne à son ID_BVOTE), alors on peut compter le nb. de lignes par arrondissement ce qui va nous donner le nb. de bureaux de vote par arrondissement 
display(data_ar)
data_ar.value_counts()
```

```python
# 4.nb. de voix dans chaque bureaux de vote obtenus par Benoît HAMON

voix_B = data_presidentielles[['HAMON Benoît']]
display(voix_B)
```

```python
# 5.nb. de voix par arrondissement

num_arr = data_presidentielles.groupby('NUM_ARROND')[['DUPONT-AIGNAN Nicolas', 'LE PEN Marine', 'MACRON Emmanuel',
       'HAMON Benoît', 'ARTHAUD Nathalie', 'POUTOU Philippe',
       'CHEMINADE Jacques', 'LASSALLE Jean', 'MÉLENCHON Jean-Luc',
       'ASSELINEAU François', 'FILLON François']].sum() # somme les nb. de voix par num. d'arrondissement

print(num_arr)

```

```python
# 6.camembert 

for i in range(1,21):   # camembert
    arr = i 
    data_arr = (data_presidentielles[data_presidentielles['NUM_ARROND'] == arr][['LE PEN Marine','MACRON Emmanuel','MÉLENCHON Jean-Luc','FILLON François']]).sum()    
    print(data_arr) # data_arr renvoie le nb. de voix pour les 4 candidats pour l'arrondissement 'arr'
    plt.figure()
    plt.pie(data_arr,labels = data_arr.index, autopct = "%1.1f%%")
    plt.title(f"arrondissement {arr}")
    plt.show()


```
