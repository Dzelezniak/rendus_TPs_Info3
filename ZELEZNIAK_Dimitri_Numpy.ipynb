```python
import numpy as np
```

```python
# création de la marche aléatoire 1D

def marche_aléatoire(p,N):  # avec 0<p<1 et N entier naturel
   
    U = np.random.random(N)  # ici on fait avec une uniforme continue qui renvoie des valeurs entre 0 et 1 
    B = (U <= p).astype(int)   # en comparant une uniforme continue Y de [0,1] à une proba 'p' cela revient à simuler une Bernoulli X de paramètre 'p' (car P(Y=<p) = p = P(X=1))
    d = np.cumsum(B)[N-1]
    g = N - d
    position = d - g
    
    return (position)
```

```python
# tests de marche_aléatoire

print(marche_aléatoire(0.5,100))
print(marche_aléatoire(0.2,100))
print(marche_aléatoire(0.8,100))
```

```python
# avec MonteCarlo

def marche_aléatoire_MC(p,N,M):  # avec 0<p<1 et N entier naturel
   
   m = 0
   s = 0
   A = []   # liste où on va sauvegarder les valeurs des M marches aléatoires
   for i in range(M):
     A.append(marche_aléatoire(p,N))
   
   for i in range(M):
     m = m + A[i]
   mean = m/M   # moyenne
   
   for i in range(M):
     s  = s + (A[i] - mean)**2
   std = np.sqrt(s/M)   # ecart-type

   return(np.array([mean,std]))
```

```python
# test avec MonteCarlo

print(marche_aléatoire_MC(0.6,100,10000)) # ici, en théorie, mean = 20 et std = sqrt(96)
print(marche_aléatoire_MC(0.2,100,10000)) # ici, en théorie, mean = -60 et std = sqrt(64) (ie. std = 8)
```

```python
# tableau pour M = 10, 100, 1 000 et 10 000

def tableau(p,N):

    tab = marche_aléatoire_MC(p,N,10)
    for i in range(3):
        tab  = np.vstack([tab, marche_aléatoire_MC(p,N,10**(i+2))])
    tabfinal = np.transpose(tab)
    return (tabfinal)

print(tableau(0.2,100))
    
```
