```python
# importation des bibliothèques 

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
# on importe la base de données 'olympics.csv'

OG = pd.read_csv("olympics.csv", sep="\t")
OG.columns
```

```python
# participation par saison

sns.set()  # utile si modifications des paramètres du graphique (# 'chart' en anglais)
chart = sns.catplot(x = "Year", data = OG, aspect = 2, kind = "count", palette = ["m", "g"])  # affiche le nb. d'athlètes/participants(# kind = "count" compte le nb de lignes par année) par année   
chart.set_xticklabels(step = 8) # affiche la graduation en abscisse des années avec un pas de 8 (ie. affiche la graduation toutes les 8 saisons de JO)
```
